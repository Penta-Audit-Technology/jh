## 최초의 부실징후데이터 만들기

#### 데이터 불러오기

```python
import pandas as pd
import numpy as np
# pd.options.display.max_columns = None
df = pd.read_csv('./csv/확인용입니다v2.csv', encoding='cp949')
df['이자보상배율'] = df['이자보상배율'].replace(0,1)
df['회계년도'] = pd.to_datetime(df['회계년도'], format='%d-%b').dt.strftime('20%d-%m')
df['거래소코드'] = df['거래소코드'].astype(str).str.zfill(6)
df
```

#### 이자보상배율 값 채우기

#### 이자보상배율 결측치를 임의의값으로 채운다

```python
df['이자보상배율'] = df['이자보상배율'].fillna(12547896.63)
```

#### 영업손익이 양수인 값(7777)을 기업의 max값으로 치환한다

```python
df['이자보상배율'] = df['이자보상배율'].replace(7777,np.nan)
maxi = lambda x: x.fillna(x.max())
df['이자보상배율'] = df['이자보상배율'].groupby(df['회사명']).apply(maxi)
df.isna().sum()
```

#### 똑같이 반복한다

```python
df['이자보상배율'] = df['이자보상배율'].fillna(12547896.63)
```

#### 영업손익이 음수인 값(-7777)을 기업의 min값으로 치환한다

```python
mins = lambda x: x.fillna(x.min())
df['이자보상배율'] = df['이자보상배율'].replace(-7777,np.nan)
df['이자보상배율'] = df['이자보상배율'].groupby(df['회사명']).apply(mins)
df.isna().sum()
```

#### 임의의 값을 다시 결측치로 변환한다

```python
df['이자보상배율'] = df['이자보상배율'].replace(12547896.63,np.nan)
df.isna().sum()
```

### 부실징후 기업 나누기

```python
# 현재 데이터 기준으로 한계기업과 4년 연속 이.보.배 1 미만인 기업을 찾을 겁니다

f_company = [] # 한계기업을 저장하는 리스트
f_company_x = [] # 4년 연속 이자보상배율 1 미만인 기업 삽입
f_safe = []
fs_company = []
count = 0 # 한계기업 카운트용
s_count = 0
company = 'x' # 한계기업 기록용
s_company = 'o'

# 이거도 csv 한바퀴 돌립니다
for i in range(0, len(df)-1):

    # 회사명이 이미 기록된 상태에서 또 중복으로 적을 필요는 없습니다
    if company == df.loc[i]['회사명']:
        # 다만 그 기업이 한계기업인데
        if count > 1:
            # 어머? 4년 연속이네요? 기록!!!
            if df.iloc[i]['이자보상배율'] < 1:
                f_company_x.append(company)
                count = 0
                # 반복문에서 볼일 다 끝났으니 통과합니다
            continue
        
        # 그게 아니라면 반복문을 통과합니다
        continue

    # 이자보상배율이 1 미만인 경우 카운트를 늘립니다
    if df.iloc[i]['이자보상배율'] < 1:
            count += 1
    # 도중에 이자보상배율이 1 이상이면 탈출합니다
    else:
        count = 0
    # 한계 기업 클럽에 오셨습니다. 환영합니다.
    if count == 1:
        company = df.loc[i]['회사명']
        f_company.append(company)
        count += 1
        
    # else: 
    #     if count == 7:
    #         company == df.loc[i]['회사명']
    #         f_safe.append(company)
    #         count += 1
    
```

```python
df_5 = [] # 임시 저장용 리스트

# 그렇게 만든 데이터프레임을 기존과 비교해서 True인 애들만 뽑아옵니다
for i in range(0, len(df_3)-1):
    if df_3.iloc[i]["회사명"] == True:
        df_5.append(df.loc[i])
        
df_5_frame = pd.DataFrame(df_5).reset_index(drop=True)
```

```python
# 4년 연속 이자 보상 배율이 1 미만인 기업들도 똑같이 해봅니다

df_6 = df.isin(f_company_x)
df_7 = []

for i in range(0, len(df_6)-1):
    if df_6.iloc[i]["회사명"] == True:
        df_7.append(df.loc[i])

df_7_frame = pd.DataFrame(df_7).reset_index(drop=True)
```

### 거래소코드 9인 외국기업 제거하기

#### 기업 기록이 2개 이하인 기업 제거

```python
# # 기업 기록이 2개 이하인 기업
single_rows = dd['회사명'].value_counts() <= 2
single_rows[single_rows]
```
