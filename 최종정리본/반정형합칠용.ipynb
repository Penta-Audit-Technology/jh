### 반정형 합칠 데이터 만들기

#### 최종데이터와 반정형데이터 불러오기

```python
import pandas as pd
import numpy as np


df = pd.read_csv('./csv/최종1년부실징후데이터(전)v3.csv',encoding='cp949')
df1 = pd.read_csv('./csv/세세분류.csv',encoding='cp949')
df2 = pd.read_csv('./csv/반정형합쳐야되는거.csv',encoding='cp949')
df = df.drop(columns=('통계청 한국표준산업분류 10차(소분류)'))
df['거래소코드'] = df['거래소코드'].astype(str).str.zfill(6)
df1['거래소코드'] = df1['거래소코드'].astype(str).str.zfill(6)
df2['거래소코드'] = df2['거래소코드'].astype(str).str.zfill(6)
df
```

### 반정형데이터 회계년도 맞추기

```python
df2['회계년도'] = pd.to_datetime(df2['회계년도'],format='%d-%b').dt.strftime('20%d-%m')
df2 = df2.drop(columns=['0','회사명'],axis=1)
df2
```

```python
df2.isna().sum()
```

### 결측치는 기업의 평균값으로 대체 후 남은 값은 0으로 처리한다

```python
men = lambda x: x.fillna(x.mean())
df = df.groupby('회사명').apply(men)
df = df.fillna(0)
df
```

```python
df.isna().sum()
```

#### 세세분류 회계년도 맞추기

```python
df1['회계년도'] = pd.to_datetime(df1['회계년도'],format='20%d/%m').dt.strftime('20%d-%m')
df1
```

### 데이터 합치기

```python
df2 = pd.merge(df,df2, on=['거래소코드','회계년도'], how='outer')
df2 = df2.dropna()
df2
```

```python
df3 = pd.merge(df2,df1, on=['회사명','거래소코드','회계년도'], how='outer')
df3 = df3.dropna()
df3
```

### 추가적인 수명주기 반정형 데이터 합치기

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./csv/1차최종데이터.csv',encoding='cp949')
df['회계년도'] = pd.to_datetime(df['회계년도'],format='20%d-%m').dt.strftime('20%d%m')
df1 = pd.read_csv('./csv/수명주기.csv' ,encoding='cp949')
df1['회계년도'] = pd.to_datetime(df1['회계년도'],format='20%d/%m').dt.strftime('20%d%m')
df1 = df1[['회사명','거래소코드','회계년도','기업수명주기']]
df
```

```python
df1
```

```python
df2 = pd.merge(df,df1, on=['회사명','거래소코드','회계년도'],how='outer')
df2 = df2.dropna()
df2 = df2.reset_index(drop=True)
df2
```

#### 수명주기 0값이나 결측치 처리하기

```python
for i in range(len(df2)):
    if df2.iloc[i]['기업수명주기'] == 9:
        if df2.iloc[i-1]['기업수명주기'] == 1:
            df2.loc[i,'기업수명주기'] = 1
        elif df2.iloc[i-1]['기업수명주기'] == 2:
            df2.loc[i,'기업수명주기'] = 2
        elif df2.iloc[i-1]['기업수명주기'] == 3:
            df2.loc[i,'기업수명주기'] = 3
        elif df2.iloc[i-1]['기업수명주기'] == 4:
            df2.loc[i,'기업수명주기'] = 4
        elif df2.iloc[i-1]['기업수명주기'] == 5:
            df2.loc[i,'기업수명주기'] = 5
```

```python
df2.to_csv('수명주기붙힌거v3.csv'encoding='cp949',index=False)
```
