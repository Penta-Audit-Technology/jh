```python
import pandas as pd
import numpy as np

df = pd.read_csv('./비율/영업이익계산용.csv',encoding='cp949')
df.columns = df.columns.str.replace('\(', '')
df.columns = df.columns.str.replace('\)', '')
df.columns = df.columns.str.replace('\*', '')
df.columns = df.columns.str.replace('IFRS', '')
df.columns = df.columns.str.replace('천원', '')
df
```
