```python
import pandas as pd
import numpy as np

df = pd.read_csv('../main/요게최종본.csv',encoding='cp949')
df['회계년도'] = pd.to_datetime(df['회계년도'],format='%d-%b').dt.strftime('20%d%m')
df1 = pd.read_csv('../수명주기.csv' ,encoding='cp949')
df1['회계년도'] = pd.to_datetime(df1['회계년도'],format='20%d/%m').dt.strftime('20%d%m')
df1 = df1[['회사명','거래소코드','회계년도','기업수명주기']]
df
```

```python
df1
```

```python

df2 = pd.merge(df,df1, on=['회사명','거래소코드','회계년도'],how='outer')
df2 = df2.dropna()
df2 = df2.reset_index(drop=True)
df2
```

```python
df2[df2['기업수명주기']==9]
```

```python
df2.to_csv('수명주기붙힌거.csv',encoding='cp949',index=False)
```

```python
for i in range(len(df2)):
    if df2.iloc[i]['기업수명주기'] == 9:
        if df2.iloc[i-1]['기업수명주기'] == 1:
            df2.loc[i,'기업수명주기'] = 1
        elif df2.iloc[i-1]['기업수명주기'] == 2:
            df2.loc[i,'기업수명주기'] = 2
        elif df2.iloc[i-1]['기업수명주기'] == 3:
            df2.loc[i,'기업수명주기'] = 3
        elif df2.iloc[i-1]['기업수명주기'] == 4:
            df2.loc[i,'기업수명주기'] = 4
        elif df2.iloc[i-1]['기업수명주기'] == 5:
            df2.loc[i,'기업수명주기'] = 5
```
