## 상관분석 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False)
```

### 데이터불러오기

```python
df = pd.read_csv('./csv/00_증가율1차.csv',encoding='cp949')
df
```

### 상관계수 만들기

```python
# 피어슨 상관계수로 상관계수를 계산합니다.
df_corr = df.corr(method='pearson')
df_corr
```

```python
df_corr['총자본증가율'].sort_values()
```

- 0에 가까운 값이면 상관관계가 없다고 볼 수 있다.

```python
# 상관계수가 0.3이상인 feature만 가져옵니다.
df_corr.loc[df_corr['총자본증가율'] >0.3,'총자본증가율']
```

```python
# 그냥 상관없이 0.3이상인 feature만 가져옵니다.
df_corr.loc[df_corr['노동장비율'] >0.1,'노동장비율']
```

```python
df_corr['노동장비율'].sort_values()
```

```python
# 상관계수를 내림차순으로 정렬합니다.
df_corr['순이익증가율'].sort_values(ascending=False).head(7)
```

### 히트맵
- 그래프와 달리 시각적으로 볼때
- 상관 관계가 클수록 색이 진해집니다

```python
# annot 숫자를 나타낼수 있고, fmt는 소수점 갯수를 지정, cmap은 색상을 지정
plt.figure(figsize=(40,20))
sns.heatmap(df_corr, annot=True, fmt=".2f", cmap="Blues")
```

### 히트맵 대각선 아래만 표시하는 방법

```python
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))
plt.figure(figsize=(40,20))
sns.heatmap(df_corr, annot=True, fmt=".2f", cmap="Blues", mask=mask)
```
