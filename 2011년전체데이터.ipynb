```python
df['한계기업'] = 0
for i in range(len(df)-2): # 3년 연속 부실징후가 1일 때
    if df['부실징후기업'][i] == 1:
        if df.loc[i,'회사명'] == df.loc[i+1,'회사명']:
            if df['부실징후기업'][i+1] == 1:
                if df['부실징후기업'][i+2] == 1:
                    df['한계기업'][i+2] = 1
        if df.loc[i,'회사명'] == df.loc[i-1,'회사명']:
            if df['부실징후기업'][i] == 1:
                if df['부실징후기업'][i+1] == 1:
                    if df['부실징후기업'][i+2] == 1:
                        df['한계기업'][i+2] = 1
        elif df.loc[i,'회사명'] != df.loc[i-1,'회사명']:
            if df['부실징후기업'][i] == 1:
                df['한계기업'][i] = 0
            else:
                df['한계기업'][i] = 0
        if df.loc[i,'회사명'] != df.loc[i-1,'회사명'] and df.loc[i,'회사명'] != df.loc[i-2,'회사명']:
            
    if df['부실징후기업'][i] == 2:
        df['한계기업'][i] = 2
df
```
