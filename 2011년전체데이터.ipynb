```python
import pandas as pd
import numpy as np

df = pd.read_csv('./csv/2011전체데이터.csv', encoding='cp949')
df
```

```python
df.isna().sum()
df = df.fillna(0)
```

```python
df_1 = df[['회사명','거래소코드','회계년도','이자보상배율(이자비용)']]
df_2 = df[['회사명','거래소코드','회계년도','이자보상배율(이자비용)(IFRS연결)']]
```

```python
for i in range(len(df_1)):
    if df_1.loc[i,'이자보상배율(이자비용)'] == 0: # 두개 다 결측치 인 경우 일단 2로...
        df_1.loc[i, '부실징후기업'] = 2
    elif df_1.loc[i,'이자보상배율(이자비용)'] < 1: # GAAP만 고려할 경우
        df_1.loc[i,'부실징후기업'] = 1
    elif df_1.loc[i,'이자보상배율(이자비용)'] > 1: # GAAP만 고려할 경우
        df_1.loc[i,'부실징후기업'] = 0

df_1 # 999999999는 일단 생각하지 않은 채로 진행합니다.
```

```python
for i in range(len(df_2)):
    if df_2.loc[i,'이자보상배율(이자비용)(IFRS연결)'] == 0: # 두개 다 결측치 인 경우 일단 2로...
        df_2.loc[i, '부실징후기업'] = 2
    elif df_2.loc[i,'이자보상배율(이자비용)(IFRS연결)'] < 1: # GAAP만 고려할 경우
        df_2.loc[i,'부실징후기업'] = 1
    elif df_2.loc[i,'이자보상배율(이자비용)(IFRS연결)'] > 1: # GAAP만 고려할 경우
        df_2.loc[i,'부실징후기업'] = 0

df_2 # 999999999는 일단 생각하지 않은 채로 진행합니다.
```

```python
for i in range(len(df_1)-2): # 3년 연속 부실징후가 1일 때
    if df_1.loc[i,'부실징후기업'] == 1:
        if df_1.loc[i+1,'부실징후기업'] == 1:
            if df_1.loc[i+2,'부실징후기업'] == 1:
                df_1.loc[i+2,"한계기업"] = 1
            else:
                df_1.loc[i+2,'한계기업'] = 0
        else:
            df_1.loc[i+1,'한계기업'] = 0
    else:
        df_1.loc[i,'한계기업'] = 0
df_1
```

```python
for i in range(len(df_2)-2): # 3년 연속 부실징후가 1일 때
    if df_2.loc[i,'부실징후기업'] == 1:
        if df_2.loc[i+1,'부실징후기업'] == 1:
            if df_2.loc[i+2,'부실징후기업'] == 1:
                df_2.loc[i+2,"한계기업"] = 1
            else:
                df_2.loc[i+2,'한계기업'] = 0
        else:
            df_2.loc[i+1,'한계기업'] = 0
    else:
        df_2.loc[i,'한계기업'] = 0
df_2
```
