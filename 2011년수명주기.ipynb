```python
import pandas as pd
import numpy as np

df = pd.read_csv('./csv/전체데이터_2011_KOSPI+KOSDAQ_수명주기.csv', encoding='cp949')
df
```

```python
for i in range(len(df)):
    if df.loc[i,'영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)'] == 0:
        df.loc[i, '영업활동'] = 9
    elif df.loc[i,'영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)'] < 0:
        df.loc[i, '영업활동'] = 0
    elif df.loc[i,'영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)'] > 0:
        df.loc[i, '영업활동'] = 1
    else:
        df.loc[i, '영업활동'] = 8
        
    if df.loc[i,'투자활동으로 인한 현금흐름(*)(IFRS)(천원)'] == 0:
        df.loc[i, '투자활동'] = 9
    elif df.loc[i,'투자활동으로 인한 현금흐름(*)(IFRS)(천원)'] < 0:
        df.loc[i, '투자활동'] = 0
    elif df.loc[i,'투자활동으로 인한 현금흐름(*)(IFRS)(천원)'] > 0:
        df.loc[i, '투자활동'] = 1
    else:
        df.loc[i, '투자활동'] = 8
        
    if df.loc[i,'재무활동으로 인한 현금흐름(*)(IFRS)(천원)'] == 0:
        df.loc[i, '재무활동'] = 9
    elif df.loc[i,'재무활동으로 인한 현금흐름(*)(IFRS)(천원)'] < 0:
        df.loc[i, '재무활동'] = 0
    elif df.loc[i,'재무활동으로 인한 현금흐름(*)(IFRS)(천원)'] > 0:
        df.loc[i, '재무활동'] = 1
    else:
        df.loc[i, '재무활동'] = 8
    
    if df.loc[i,'영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'] == 0:
        df.loc[i, '영업활동(연결)'] = 9
    elif df.loc[i,'영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'] < 0:
        df.loc[i, '영업활동(연결)'] = 0
    elif df.loc[i,'영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'] > 0:
        df.loc[i, '영업활동(연결)'] = 1
    else:
        df.loc[i, '영업활동(연결)'] = 8
        
    if df.loc[i,'투자활동으로 인한 현금흐름(*)(IFRS연결)(천원)'] == 0:
        df.loc[i, '투자활동(연결)'] = 9
    elif df.loc[i,'투자활동으로 인한 현금흐름(*)(IFRS연결)(천원)'] < 0:
        df.loc[i, '투자활동(연결)'] = 0
    elif df.loc[i,'투자활동으로 인한 현금흐름(*)(IFRS연결)(천원)'] > 0:
        df.loc[i, '투자활동(연결)'] = 1
    else:
        df.loc[i, '투자활동(연결)'] = 8
        
    if df.loc[i,'재무활동으로 인한 현금흐름(*)(IFRS연결)(천원)'] == 0:
        df.loc[i, '재무활동(연결)'] = 9
    elif df.loc[i,'재무활동으로 인한 현금흐름(*)(IFRS연결)(천원)'] < 0:
        df.loc[i, '재무활동(연결)'] = 0
    elif df.loc[i,'재무활동으로 인한 현금흐름(*)(IFRS연결)(천원)'] > 0:
        df.loc[i, '재무활동(연결)'] = 1
    else:
        df.loc[i, '재무활동(연결)'] = 8
df
```

```python
df_1 = df.copy()
for i in range(len(df_1)):
    # 도입기
    if df_1.loc[i,'영업활동'] == 0 and df_1.loc[i,'투자활동'] == 0 and df_1.loc[i,'재무활동'] == 1:
        df_1.loc[i,'기업수명'] = 1
    # 성장기
    elif df_1.loc[i,'영업활동'] == 1 and df_1.loc[i,'투자활동'] == 0 and df_1.loc[i,'재무활동'] == 1:
        df_1.loc[i,'기업수명'] = 2
    # 성숙기
    elif df_1.loc[i,'영업활동'] == 1 and df_1.loc[i,'투자활동'] == 0 and df_1.loc[i,'재무활동'] == 0:
        df_1.loc[i,'기업수명'] = 3
    # 쇠퇴기
    elif df_1.loc[i,'영업활동'] == 0 and df_1.loc[i,'투자활동'] == 1 and df_1.loc[i,'재무활동'] == 1:
        df_1.loc[i,'기업수명'] = 5
    elif df_1.loc[i,'영업활동'] == 0 and df_1.loc[i,'투자활동'] == 1 and df_1.loc[i,'재무활동'] == 0:
        df_1.loc[i,'기업수명'] = 5
    # 결측치인 경우
    elif df_1.loc[i,'영업활동'] == 8 or df_1.loc[i,'투자활동'] == 8 or df_1.loc[i,'재무활동'] == 8:
        df_1.loc[i,'기업수명'] = 8
    elif df_1.loc[i,'영업활동'] == 8 and df_1.loc[i,'투자활동'] == 8 and df_1.loc[i,'재무활동'] == 8:
        df_1.loc[i,'기업수명'] = 8
    # 0인 경우
    elif df_1.loc[i,'영업활동'] == 9 or df_1.loc[i,'투자활동'] == 9 or df_1.loc[i,'재무활동'] == 9:
        df_1.loc[i,'기업수명'] = 9
    elif df_1.loc[i,'영업활동'] == 9 and df_1.loc[i,'투자활동'] == 9 and df_1.loc[i,'재무활동'] == 9:
        df_1.loc[i,'기업수명'] = 9
    # 나머지는 쇠신기라고 할 경우
    else:
        df_1.loc[i,'기업수명'] = 4
        
    # 연결일 경우
        
    # 도입기
    if df_1.loc[i,'영업활동(연결)'] == 0 and df_1.loc[i,'투자활동(연결)'] == 0 and df_1.loc[i,'재무활동(연결)'] == 1:
        df_1.loc[i,'기업수명(연결)'] = 1
    # 성장기
    elif df_1.loc[i,'영업활동(연결)'] == 1 and df_1.loc[i,'투자활동(연결)'] == 0 and df_1.loc[i,'재무활동(연결)'] == 1:
        df_1.loc[i,'기업수명(연결)'] = 2
    # 성숙기
    elif df_1.loc[i,'영업활동(연결)'] == 1 and df_1.loc[i,'투자활동(연결)'] == 0 and df_1.loc[i,'재무활동(연결)'] == 0:
        df_1.loc[i,'기업수명(연결)'] = 3
    # 쇠퇴기
    elif df_1.loc[i,'영업활동(연결)'] == 0 and df_1.loc[i,'투자활동(연결)'] == 1 and df_1.loc[i,'재무활동(연결)'] == 1:
        df_1.loc[i,'기업수명(연결)'] = 5
    elif df_1.loc[i,'영업활동(연결)'] == 0 and df_1.loc[i,'투자활동(연결)'] == 1 and df_1.loc[i,'재무활동(연결)'] == 0:
        df_1.loc[i,'기업수명(연결)'] = 5
    # 결측치인 경우
    elif df_1.loc[i,'영업활동(연결)'] == 8 or df_1.loc[i,'투자활동(연결)'] == 8 or df_1.loc[i,'재무활동(연결)'] == 8:
        df_1.loc[i,'기업수명(연결)'] = 8
    elif df_1.loc[i,'영업활동(연결)'] == 8 and df_1.loc[i,'투자활동(연결)'] == 8 and df_1.loc[i,'재무활동(연결)'] == 8:
        df_1.loc[i,'기업수명(연결)'] = 8
    # 0인 경우
    elif df_1.loc[i,'영업활동(연결)'] == 9 or df_1.loc[i,'투자활동(연결)'] == 9 or df_1.loc[i,'재무활동(연결)'] == 9:
        df_1.loc[i,'기업수명(연결)'] = 9
    elif df_1.loc[i,'영업활동(연결)'] == 9 and df_1.loc[i,'투자활동(연결)'] == 9 and df_1.loc[i,'재무활동(연결)'] == 9:
        df_1.loc[i,'기업수명(연결)'] = 9
    # 나머지는 쇠신기라고 할 경우
    else:
        df_1.loc[i,'기업수명(연결)'] = 4
df_1
```
