```python
cols = []

total_df1_iqr = dd.copy()
total_df1_iqr = total_df1_iqr[['총자본증가율','유형자산증가율','유동자산증가율','재고자산증가율','자기자본증가율','매출액증가율','순이익증가율']]
cols.append([col for col in total_df1_iqr])

Q1 = total_df1_iqr[['총자본증가율','유형자산증가율','유동자산증가율','재고자산증가율','자기자본증가율','매출액증가율','순이익증가율']].quantile(q=0.25)

Q3 = total_df1_iqr[['총자본증가율','유형자산증가율','유동자산증가율','재고자산증가율','자기자본증가율','매출액증가율','순이익증가율']].quantile(q=0.75)

IQR = Q3 - Q1
for i in cols:
    total_df1_iqr[i] = total_df1_iqr[i][(total_df1_iqr[i] <= Q3[i]+1.5*IQR[i]) & (total_df1_iqr[i] >= Q1[i]-1.5*IQR[i])]
```

```python
# 이상치 박스플롯
from matplotlib.pyplot import imread

fig, axes = plt.subplots(2, 3)
fig.set_size_inches((20,15))

total_df1_iqr.boxplot(column="총자본증가율", ax=axes[0,0])
total_df1_iqr.boxplot(column="유형자산증가율", ax=axes[0,1])
total_df1_iqr.boxplot(column="유동자산증가율", ax=axes[0,2])
total_df1_iqr.boxplot(column="재고자산증가율", ax=axes[1,0])
total_df1_iqr.boxplot(column="자기자본증가율", ax=axes[1,1])
total_df1_iqr.boxplot(column="매출액증가율", ax=axes[1,2])
total_df1_iqr.boxplot(column="순이익증가율", ax=axes[2,1])
total_df1_iqr.boxplot(column="유형자산회전율", ax=axes[2,2])

plt.show()
```
